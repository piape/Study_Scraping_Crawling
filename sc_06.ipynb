# n개의 페이지 함수 스타일로 변환.
- 함수 이용. n은 매개변수로 이용
- 순위. 책제목, 저자, 정가, 할인가, 책상세url, 표지url
- rank title author, o_price, d_price, detail, cover
- yes24(10) 호출하면, 1~10 page 데이타가 데이타프레임으로 출력되도록 한다.

```python
# 2) 데이타프레임으로 반환되게 함수 생성
# 함수 정의
# n은 페이지 끝번호
def yes24(n):
    book_list = []    
    
    for i in range(1, n+1):
        base_url_2 = 'http://www.yes24.com/24/category/bestseller?CategoryNumber=001&sumgb=06&fetchSize=40&PageNumber='
        url = base_url_2 + str(i)
        print(i , '페이지를 데이터프레임으로 저장 중 입니다.')
        # 웹페이지 접근 및 페이지 추출
        res = requests.get(url)
        html_str = res.text
        soup = BeautifulSoup(html_str)
        target = soup.select('tr:nth-child(odd)')
        
        for i in target:
            rank = i.select_one('td.num').text.replace('.', '위')
            title = i.select_one('img')['alt']
            author = i.select_one('div.aupu a').text
            o_price = i.select('p')[1].text.split(' ')[0]
            d_price = i.select('p > span')[0].text
            detail = i.select('a > img')[0].get('src').replace('S','XL')
            cover = base_url + i.select('div a')[2].get('href')

            book_list.append([rank, title, author, o_price, d_price, detail, cover])
            
    df = pd.DataFrame(book_list, columns = ['rank', 'title', 'author', 'o_price', 'd_price', 'detail', 'cover'])
    
    return df 
```

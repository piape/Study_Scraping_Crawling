# 퀴즈1 - 다나와
https://www.danawa.com

검색어 : '삼성전자 갤럭시 워치 액티브2'

검색어로 검색된 모든 상품 정보를 데이타프레임으로 변경 확인한 후 csv 파일로 저장한다.

스펙이 없거나 평점이 없는 경우에는 공백 처리

제품명	링크	가격	평점	등록일	스펙

```python
# 모듈 임포트
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
# 셀레니움 실행, 검색
driver = webdriver.Chrome('chromedriver.exe')
time.sleep(0.5)

url = 'https://www.danawa.com'
driver.get(url)

time.sleep(0.5)

driver.find_element_by_name('k1').send_keys("삼성전자 갤럭시 워치 액티브2")
driver.find_element_by_name('k1').submit()

time.sleep(0.5)
```

```python
# 뷰티풀숲
html = driver.page_source
soup = BeautifulSoup(html)
# div tag 추출
target = soup.select('div.prod_main_info')
print(' 제품수 :', len(target))
```

```python
target[0].text.replace('\n','').replace('\t','')
```

```python
# 요소 추출
# 제품명 링크 가격 평점 등록일 스펙
# 샘플 저장 확인

prod_name = target[0].select_one('p.prod_name > a').text
prod_link = target[0].select_one('p.prod_name > a')['href']
prod_date = target[0].select_one('dl.meta_item.mt_date').text
prod_price = target[0].select_one('p.price_sect strong').text
prod_score = target[0].select_one('div.point_num strong').text
prod_spec = target[0].select_one('dd div.spec_list').text.replace('\n','').replace('\t','')

print(prod_name, prod_link, prod_price, prod_score, prod_date, prod_spec)
print('-' * 100)
```

# 1페이지 결과물 저장

```python
# 빈리스트 생성
result_list = []

# for 문
for i in target:
    try : 
        prod_name = i.select_one('p.prod_name a').text
        prod_link = i.select_one('p > a')['href']
        prod_date = i.select_one('dl.meta_item.mt_date dd').text.replace('\n','').replace('\t','')

        if i.select_one('p.price_sect strong'):
            prod_price = i.select_one('p.price_sect strong').text
        else:
            prod_price = ''

        if i.select_one('div.point_num strong'):
            prod_score = i.select_one('div.point_num strong').text
        else:
            prod_score = ''

        if i.select_one('dd div.spec_list'):
            prod_spec = i.select_one('dd div.spec_list').text.replace('\n','').replace('\t','')
        else:
            prod_spec = ''
            
        result_list.append([prod_name, prod_link, prod_price, prod_score, prod_date, prod_spec])
        
    except:
        pass
```

```python
# 출력확인
print(result_list[0]) # 1번
print(result_list[-1]) # 끝번
```

```python
# 데이타 프레임
df = pd.DataFrame(result_list, columns = ['제품명', '링크' ,'가격', '평점' ,'등록 월' ,'스펙'])
```

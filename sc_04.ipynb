# Yes24 베스트셀러 정보 저장하기

http://www.yes24.com/Main/default.aspx
- 1-10위까지 베스트셀러(순위, 책제목, 저자, 출판사, 표지, 책상세, URL)


```python
# (1) 웹페이지에서 정보 가져오기
```

```python
url = 'http://www.yes24.com/Main/default.aspx'
res = requests.get(url)
html_str = res.text
res
```

```python
# (2) bs4 적용
# html 페이지 > 숲개체

soup = bs(html_str, 'html.parser')
# soup
```

```python
# (3) 소스 파악하기

target = soup.select('li.tp02')
target

# (4) 개별확인
target[0]
```

```python
# (5) 1위에서 개별정보 추출
```

```python
# 순위, 책제목, 저자, 출판사, 표지url, 상세url
print('순위 => ', target[0].select_one('strong').text)

# 2번째 스트롱 태그에서 텍스트만  : 책제목
print('책 제목 => ', target[0].select('strong')[1].text)

# 저자
print('저자 => ', target[0].select('em')[1].text)

# 출판사 split을하면 list로 쪼개진다.
print('출판사 => ', target[0].select_one('a')['title'].split(' ')[3])

# 표지url (작은표지 이미지  -> 큰표지이미지 경로로)
print('표지URL => ', target[0].select_one('img')['src'].replace('S', 'XL'))

base_url = 'http://www.yes24.com'

# 상세url
print('상세URL => ', base_url + target[0].select_one('a')['href'])

print('-' * 100)
```

```python
# (6) 2차원 리스트로 저장

bestbook_list = []

for li in target:
    rank = li.select_one('strong').text
    title = li.select('strong')[1].text
    author = li.select('em')[1].text
    pub = li.select_one('a')['title'].split(' ')[3]
    cover = li.select_one('img')['src'].replace('S', 'XL')
    detail = base_url + li.select_one('a')['href']
    
    
    print(rank, title, author, pub, cover, detail)
    
    # 리스트에 추가
    bestbook_list.append([rank, title, author, pub, cover, detail])
    
        
```

```python
# 2차원 리스트 출력
for rank, title, author, pub, cover, detail in bestbook_list:
    print(rank, title, author, pub, cover, detail)
    
bestbook_list
```

```python
# 2차원 리스트를 DataFrame 으로 변환

# 임포트 판다스
import pandas as pd

df = pd.DataFrame(bestbook_list, columns =['rank', 'title', 'author', 'pub', 'cover', 'detail'])
```

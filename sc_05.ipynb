```python
# 2 for ~ for 문 형태로 명령 실행


# 고정 주소
base_url = 'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=167697&target=after&page='

# 페이지 끝번호
n = 5 

# 빈 리스트 생성
movie_list2 = []


for i in range(1, n+1):
    # 1) 웹피이지 접근 > 리스폰스 객체 > html 페이지
    url = base_url+str(i)
    print(url)
    res = requests.get(url)
    html_str = res.text
    
    # 솝 객체
    soup = BeautifulSoup(html_str)
    
    # tr 추출
    target = soup.select('tbody > tr')
    
    # 리스트의 추가
    for tr in target:
        title = tr.select_one('a').text
        comment = tr.select_one('br').next_sibling.strip()
        score = tr.select_one('em').text
        author = tr.select_one('a.author').text
        date = tr.select('td')[-1].select_one('br').next_sibling
        # 리스트 안의 리스트 구조로 데이타 추가 
        movie_list2.append([title, comment, score, author, date])

len(movie_list2)
```

```python
# n페이지의 마지막 게시글
movie_list2[-1]
```

```python
# 전체 데이타 프레임 
# df 
df = pd.DataFrame(movie_list2, columns = ['title', 'comment', 'score', 'author', 'date'])
```

```python
# 데이타프레임에서 마지막 데이타중 n개만 출력 
# 데이타프레임명.tail(n) or head(n)

df.tail(5)
```

# 전체페이지 댓글 저장
- 영화 예시 : 키드
- https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=13558&target=after&page=1

- 마지막 페이지는 15 but 기계가 스스로 판단

# 무한루프
- while True: 0빼고 나머지 숫자

    if 조건 => while문에서 빠져나가는 명령 필요
        break
- page _ nation

#### 마지막페이지 조건은?
#### 다음 페이지로 이동하는 [>] 버튼이 없음
#### [>] 소스확인

a.pg_next
<a href="페이지주소" title="다음" class="pg_next"><em>다음</em></a>

```python
# 파이썬의 True/False 조건

# True 조건 : True, 0이 아닌 숫자, 길이가 0이 아닌 문자열, 튜플, 리스트, 딕셔너리 ...
# False 조건 : False, 0, '', [], (), {}, set({}), None ...
```

```python
# 1) 마지막 페이지 판독 테스트

# 고정주소
base_url = 'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=13558&target=after&page='

# 페이지 초기값 설정
page = 1

while True:
    # 1-1) 웹페이지 접근 > 리스폰스 객체 > html 페이지 생성
    url = base_url+str(page)
    print(url)
    res = requests.get(url)
    html_str = res.text
    
    # 1-2) 솝 객체
    soup = BeautifulSoup(html_str)
    
    # 1-3) tr 추출
    target = soup.select('tbody > tr')
    
    # 1-4) 마지막 다음 버튼
    target = soup.select_one('a.pg_next')
    print('target =', target)
    
    if target:
        #페이지 번호 증가
        page += 1
    else:
        print('마지막 페이지입니다.')
        break
```

```python
# 2) 페이지 주소를 변경하면서 2차원 리스트에 데이터 저장

# 고정주소 = 키드 영화
base_url = 'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=13558&target=after&page='

# 페이지 초기값 설정
page = 1

# 빈리스트 생성
movie_list3 = []


while True:
    # 2-1) 웹페이지 접근 => response 객체 => html 페이지
    url = base_url+str(page)
    print(url)
    res = requests.get(url)
    html_str = res.text    
    # 2-2) 뷰티풀숩 객체화 : html 페이지 => 숩객체 
    soup = BeautifulSoup(html_str)
    # 2-3) 마지막 다음 버튼이 있다면 데이타 추가후 페이지 번호 증가  
    # tr 추출
    target2 = soup.select('tbody > tr')
    # 리스트에 추가 
    for tr in target2:
        title = tr.select_one('a').text
        comment = tr.select_one('br').next_sibling.strip()
        score = tr.select_one('em').text
        writer = tr.select_one('a.author').text
        date = tr.select('td')[-1].select_one('br').next_sibling        
        # 리스트 안의 리스트 구조로 데이타 추가 
        movie_list3.append([title, comment, score, writer, date])
        #페이지 번호 증가
        # 마지막 다음 버튼이 없다면 데이타 추가 while 탈출 
    target = soup.select_one('a.pg_next')
    if target:
        # 페이지 번호 증가 
        page += 1
    else: 
        print('마지막 페이지입니다.')
        break
```

```python
len(movie_list3)
```

```python
# 마지막 페이지의 마지막 게시물 확인 
movie_list3[-1]
```

```python
df = pd.DataFrame(movie_list3, columns = ['title', 'comment', 'score', 'author', 'date'])
```

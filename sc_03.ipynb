### 블로터의 실시간 랭킹 텍스트만 출력하시오.

- https://www.bloter.net/

```python
# (1) 웹페이지 접근 > response 객체 > html 소스
# 예제 소스 : 블로터
# 객체화
url = 'https://www.bloter.net/'
res = requests.get(url)
print(res)
res.text
```

```python
# 304 상태코드 => 200 상태코드 
# 사이트 이용 => http://www.useragentstring.com

# 유저정보 생성 
user_agent_info = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5005.61 Safari/537.36'

# 헤더 정보 생성 
headers = {'User-Agent' : user_agent_info}

# 객체화 
url = 'https://www.bloter.net'
res = requests.get(url, headers=headers)
print(res) 

# 객체로부터 html 페이지 생성 
html_str = res.text
html_str
```

```python
# (2) html 소스 => BeautifulSoup 객체화
# soup변수명 = BeautifulSoup(html소스, 해석기(html.parser/html5/lxml/xml))
soup = bs(html_str, 'html.parser')

print(type(soup))

soup
```

# 영문 색상명 + 16진수색상값
- 예제 사이트 https://www.w3schools.com/colors/colors_names.asp
- 영문 색상명, 16진수 색상값, 관련 url 정보 -> color.csv로 저장

```python
# (1) 웹페이지 접근 > response 객체 > html 소스
# 예제 소스 : w3 schools > html color name

# 객체화
url = 'https://www.w3schools.com/colors/colors_names.asp'
res = requests.get(url)
print(res)
res.text
```

```python
html_str = res.text
html_str
```

```python
# (2) html 소스 => BeautifulSoup 객체화
# soup변수명 = BeautifulSoup(html소스, 해석기(html.parser/html5/lxml/xml))
soup = bs(html_str, 'html.parser')
print(type(soup))
soup
```

```python
# (4) 색상 리스트 생성
color_list = soup.find_all(class_='innerbox')
# 셀렉트로 접근시
# color_list = soup.select('.innerbox')
```

```python
# 전체 개수 확인
print(len(color_list))

print('-' * 100)

print(color_list[0])
print('-' * 100)
print(color_list[-1])  # or color_list[147]
print('-' * 100)
```

```python
# (5) 첫 번째 색상 정보만 추출

# 색상명 
print('색상명 :', color_list[0].find('a').text)
# a 태그 중 2번째 텍스트
print('16 진수 색상값 :', color_list[0].find_all('a')[1].text)

# 색상 url
base_url = 'https://www.w3schools.com'
print('색상 url :', base_url + color_list[0].find('a')['href'])
```

```python
# (6) 2차원 리스트로 저장
# 빈리스트 작성

color_list2 = []
base_url = 'https://www.w3schools.com'
for color in color_list:
    print('색상명 :', color.find('a').text)
    print('16 진수 색상값 :', color.find_all('a')[1].text)
    print('색상 url :', base_url + color.find('a')['href'])
    # 리스트에 저장
    color_list2.append([color.find('a').text, color.find_all('a')[1].text, base_url + color.find('a')['href']])
```

```python
print(len(color_list2))
```

```python
# 2차원 리스트를 출력
for i, j, k in color_list2:
    print(i, j, k)
```

```python
# (7) 2차원 리스트를 csv 로 저장 > 판다스의 데이터프레임 구조
# 판다스 임포트
import pandas as pd

pd.__version__
```

```python
# 데이터프레임명 = pd.DataFrame(2차원리스트, columns=[컬럼명1 . 2 . 3 . . .])
df = pd.DataFrame(color_list2, columns=['색상명', '16진수 색상값', 'URL'])
```

```python
df
```

```python
# (8) 데이타 프레임 -> csv로 저장
# 데이타프레임.to_csv('csv저장파일경로', index=False)
df.to_csv('output/color.csv', index = False, encoding='cp949')
```
